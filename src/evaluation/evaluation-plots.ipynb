```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
def show_values(axs, orient="v", space=.01):
    def _single(ax):
        if orient == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + (p.get_height() * 0.01)
                value = '{:.1f}'.format(p.get_height())
                ax.text(_x, _y, value, ha="center")
        elif orient == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - (p.get_height() * 0.5)
                value = '{:.1f}'.format(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)
```

```python
def show_values_as_percent(axs, orient="v", space=.01):
    def _single(ax):
        if orient == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + (p.get_height() * 0.01)
                value = '{:.1f}'.format(p.get_height())
                ax.text(_x, _y, value + "%", ha="center")
        elif orient == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - (p.get_height() * 0.5)
                value = '{:.1f}'.format(p.get_width())
                ax.text(_x, _y, value + "%", ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)
```

# Compared against LongplayingPolicy

```python
against_lpp_winning_ratio_dict = {'Agent': ["Less Observation", "More Observation", "Trained against Agents"]
    , 'Winning Ratio': [66, 89.4, 90.8]}

p = sns.barplot(data=pd.DataFrame(against_lpp_winning_ratio_dict), x="Agent", y="Winning Ratio", width=0.6)

show_values_as_percent(p)
```

```python
against_average_points_dict = {'Agent': ["Less Observation", "More Observation", "Trained against Agents"]
    , 'Average Points': [23.50, 43.03, 47.49]}

p = sns.barplot(data=pd.DataFrame(against_average_points_dict), x="Agent", y="Average Points", width=0.6)
show_values(p)
```

# Compared against Random Agent

```python
against_rand_winning_ratio_dict = {'Agent': ["Less Observation", "More Observation", "Trained against Agents"]
    , 'Winning Ratio': [95.6, 99.8, 100]}

p = sns.barplot(data=pd.DataFrame(against_rand_winning_ratio_dict), x="Agent", y="Winning Ratio", width=0.6)
show_values_as_percent(p)
```

```python
against_rand_average_points_dict = {'Agent': ["Less Observation", "More Observation", "Trained against Agents"]
    , 'Average Points': [24.01, 43.23, 50.25]}

p = sns.barplot(data=pd.DataFrame(against_rand_average_points_dict), x="Agent", y="Average Points", width=0.6)
show_values(p)
```
